```python
median_price = housing.SalePrice.median()
median_price
```

```python
df1 = housing.groupby(['Neighborhood'])['SalePrice'].median().reset_index()
df1 = df1.rename(columns={'SalePrice':'median_saleprice'})
df1
```

```python
df1['index_to_median_saleprice'] = (df1['median_saleprice']/median_price*100).astype(int)
df1
```

```python
df2= housing[['Neighborhood','SalePrice']]
df2 = df2.groupby(['Neighborhood']).count().reset_index()
df2 = df2.rename(columns={'SalePrice':'num_of_houses'})
```
