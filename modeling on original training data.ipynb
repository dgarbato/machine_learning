```python
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
%matplotlib inline

from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_price = pd.read_csv(r'C:\Users\dgarb\OneDrive\Documents\Data Science Bootcamp August\Machine Learning Project\iteration files\house_price.csv',index_col=0)
```

```python
house_price.head(3)
```

```python
log_price = house_price['log_price']
```

```python
vars = mod_data.columns
vars
```

### iteration zero 

```python
it0 = ['log_GrLivArea', 'log_LotArea','log_MasVnrArea',  'yeo_TotalBsmtSF', 'yeo_GarageArea','yeo_WoodDeckSF', 'yeo_OpenPorchSF']
```

```python
ss = StandardScaler()
X = ss.fit_transform(mod_data[it0])

features = it0
###############################################
lasso = Lasso(max_iter=3000)

params = {"alpha": [0.0001, 0.001, 0.01, 0.1, 1, 10]}
gs = GridSearchCV(lasso, params, cv=5, return_train_score=True, refit=True)

#gs is now the gs object with lasso inside
gs.fit(X,log_price)
gs.best_params_
###################################    
    
print('best params: ',gs.best_params_)
print('score: ',gs.score(X,log_price))
print('  ')



adj_r2 = 1-(1-gs.score(X,log_price))*(X.shape[0] - 1) / (X.shape[0] - X.shape[1] - 1)
adj_r2
print('adj_r2: ', adj_r2)
print('  ')

lasso_pred = gs.best_estimator_.predict(X)
print('RMSE: ' + str(mean_squared_error(lasso_pred,log_price)**0.5))
print('  ')


coefs = pd.Series(gs.best_estimator_.coef_ )

varnames = pd.Series(features)



coefs_ = pd.concat([varnames, coefs], axis=1)

# print(coefs_)

type(gs.best_estimator_.coef_)
type(coefs)

# print('coefs: ',coefs_)
# print('sorted_coefs', sorted(list(zip(mod_data.columns,coefs)),reverse=True, key=lambda x: x[1]))
type(varnames)
print(coefs_.sort_values(by=[1], ascending=False))
print(' ')

import scipy

residuals = log_price - lasso_pred
print('skewness: ',scipy.stats.skew(residuals))
print(' ')


sns.set_theme(style='darkgrid')
sns.histplot(residuals, bins=20);
```

```python
ss = StandardScaler()
X = ss.fit_transform(mod_data_dum)

features = list(mod_data_dum.columns)

lasso = Lasso(max_iter=3000)

params = {"alpha": [0.0001, 0.001, 0.01, 0.1, 1, 10]}
gs = GridSearchCV(lasso, params, cv=5, return_train_score=True, refit=True)
gs.fit(X,log_price)
gs.best_params_
###################################    
    
print('best params: ',gs.best_params_)
print('score: ',gs.score(X,log_price))
print('  ')



adj_r2 = 1-(1-gs.score(X,log_price))*(X.shape[0] - 1) / (X.shape[0] - X.shape[1] - 1)
adj_r2
print('adj_r2: ', adj_r2)
print('  ')

lasso_pred = gs.best_estimator_.predict(X)
print('RMSE: ' + str(mean_squared_error(lasso_pred,log_price)**0.5))
print('  ')


coefs = pd.Series(gs.best_estimator_.coef_, name = 'coef' )

varnames = pd.Series(features, name = 'features')

pd.set_option('display.max_rows', None) # or 1000.

abs_coefs = pd.Series(np.abs(coefs),name = 'abs_val')

coefs_ = pd.concat([varnames,coefs, abs_coefs], axis=1)

# print(coefs_)

type(gs.best_estimator_.coef_)
type(coefs)

# print('coefs: ',coefs_)
# print('sorted_coefs', sorted(list(zip(mod_data.columns,coefs)),reverse=True, key=lambda x: x[1]))
type(varnames)

coefs_sorted = coefs_.sort_values(by=['coef'], ascending=False)

print(coefs_.sort_values(by=['abs_val'], ascending=False))
print(' ')

import scipy

residuals = log_price - lasso_pred
print('skewness: ',scipy.stats.skew(residuals))
print(' ')


sns.set_theme(style='darkgrid')
sns.histplot(residuals, bins=20);
```
